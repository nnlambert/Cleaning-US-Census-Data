# Cleaning US Census Data

The objective of this project is to clean this us census data and ultimately create several graphs from which informatuion can be interpreted

```python
#We can see the csv files assosicated with this project are all states + #
#We'll use .glob() to read them all in at once
```

```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob

#Use glob to load in all states csv files
files = glob.glob("states*.csv")

#Create an empty list to add read in csvs
df_list = []

#looping through csv files to read and then append them to empty list
for filename in files:
    data = pd.read_csv(filename)
    df_list.append(data)

#Joining csv files together all at once in dataframe called df
df = pd.concat(df_list)
```

```python
#Checking Dataframe for any obstacles for making a histogram
print(df.columns)
print(df.dtypes)
```

```python
#Income is shown as an object | Checking head for more info | 
print(df.head())
```

```python
#Converting Income to a numeric type for later analysis with for loop
for i in range(0, len(df["Income"])):
    str_income = str(df['Income'].iat[i])
    dollar_strip = str_income.replace('$', '')
    comma_strip = dollar_strip.replace(',', '')
    df['Income'].iat[i] = comma_strip

df['Income'] = pd.to_numeric(df['Income'])

#DF Check
df
```

```python
#Splitting GenderPop column to make male and female data points more clear
dfg = df["GenderPop"].str.split("_")
#print(dfg.head())

#Create empty list to append results of dfg to
Malep = []
Femalep = []

#Iterate through lists created in dfg 
for i in dfg:
    Malep.append(i[0])
    Femalep.append(i[1])

#Checking New Lists
#print(Malep)
#print(Femalep)

#Placing them in their own columns
df["MalePop"] = Malep
df["FemalePop"] = Femalep

#Checking Results
print(df.head())
```

7. Convert both of the columns into numerical datatypes.

   There is still an `M` or an `F` character in each entry! We should remove those before we convert.

```python
#Stripping Letter Description
df["MalePop"] = df["MalePop"].str.replace('M', '')
df["FemalePop"] = df["FemalePop"].str.replace('F', '')

#Converting to numeric values
df["MalePop"] = pd.to_numeric(df["MalePop"])
df["FemalePop"] = pd.to_numeric(df["FemalePop"])
```

```python
print(df.dtypes)
```

```python
print(df.head())
```

```python
plt.scatter(df["FemalePop"], df["Income"])
plt.title("Scatter Plot of Income vs. # of Women per State")
plt.xlabel("Population of Women per State")
plt.ylabel("Income (in US Dollars)")
plt.show()
plt.clf()
```

```python
#Fill in NaN values in women column for further analysis
df["FemalePop"] = df["FemalePop"].fillna(df["TotalPop"] - df["MalePop"])
print(df["FemalePop"])
```

```python
#Checking for duplicate data
dupes = df.duplicated(subset = df.columns[1:])
print(dupes.value_counts())
```

```python
undupe = df.drop_duplicates(subset = df.columns[1:])
print(undupe)
```

```python
plt.scatter(df["FemalePop"], df["Income"])
plt.title("Scatter Plot of Income vs. # of Women per State")
plt.xlabel("Population of Women per State")
plt.ylabel("Income (in US Dollars)")
plt.show()
plt.clf()
```

## Histogram of Races

```python
#Taking a look at the data by race
df.columns
```

```python
#Using a for loop to create plots for each race
for race in ['Hispanic', 'White', 'Black', 'Native', 'Asian', 'Pacific']:
    plt.hist(df[race])
    plt.title("Histogram of the percentage of {} People per State".format(race))
    plt.xlabel("Percentage")
    plt.ylabel("Frequency")
    plt.show()
    plt.clf()
```
